```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Model, Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import layers, models
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
```

```python
path = 'ecg.csv'
df = pd.read_csv(path,header=None)
df = df.drop(140, axis=1)
df
# for credit card
# path = 'dataset/creditcardfraud-csv/creditcardfraud-csv/creditcard.csv'
# df = pd.read_csv(path)
# df = df.drop(['Time','Class'], axis=1)
# df
```

```python
StandardScaler = StandardScaler()
df = StandardScaler.fit_transform(df)
# don't use for credit card
```

```python
x_train, x_test = train_test_split(df, test_size=0.2)
print(x_train.shape[1])
print(x_test.shape[1])
type(x_train)
```

```python
encoder = tf.keras.models.Sequential([
    layers.Input(shape=(x_train.shape[1],)),
    layers.Dense(32, activation='relu'),#64
    layers.Dense(16, activation='relu'),#64
    layers.Dense(8, activation='relu')#20
])

decoder = tf.keras.models.Sequential([
    layers.Input(shape=(8,)),#20
    layers.Dense(16, activation='relu'),#64
    layers.Dense(32, activation='relu'),#64
    layers.Dense(x_train.shape[1], activation='linear')  # Use linear activation for reconstruction
])

model = tf.keras.models.Sequential([
    encoder,
    decoder
])
```

```python
model.compile(optimizer='adam', loss ='mean_squared_error')
```

```python
history = model.fit(
    x_train,
    x_train,
    validation_data=(x_test,x_test),
    epochs=10,#5
    batch_size = 30,#100
    shuffle=True
)
```

```python
sns.lineplot(model.history.history)
```

```python
predictions = model.predict(x_test)
mse = np.mean(np.power(x_test - predictions, 2), axis=1)
```

```python
threshold = np.percentile(mse, 95)  # Adjust the percentile as needed
threshold
```

```python
anomalies = mse > threshold
```

```python
num_anomalies = np.sum(anomalies)
print(f"Number of Anomalies: {num_anomalies}")
```

```python
plt.plot(mse, marker='o', linestyle='', markersize=3, label='MSE')
plt.axhline(threshold, color='r', linestyle='--', label='Anomaly Threshold')
plt.xlabel('Sample Index')
plt.ylabel('MSE')
plt.title('Anomaly Detection Results')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x_test[0], label='Original ECG')
plt.plot(predictions[0], label='Reconstructed ECG')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Normal ECG')
plt.show()
# for credit card
# plt.figure(figsize=(12, 6))
# plt.plot(x_test.iloc[0], label='Original ECG')
# plt.plot(predictions[0], label='Reconstructed ECG')
# plt.xlabel('Time')
# plt.ylabel('Amplitude')
# plt.legend()
# plt.title('Normal ECG')
# plt.show()
```
